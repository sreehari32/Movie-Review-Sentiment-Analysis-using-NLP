```python
# removing html
from bs4 import BeautifulSoup
def remove_html(text):
    soup= BeautifulSoup(text,'lxml')
    html_free=soup.get_text()
    return html_free
```

```python
df['review']=df['review'].apply(remove_html)
```

```python
df.review[1]
```

```python
# converting to lower case
df['review']=df['review'].apply(lambda x:x.lower())
```

```python
# remove punctuations 

import string
exclude=string.punctuation


def rem_punc(text):
    text = text. translate(str.maketrans(' ', ' ' , exclude))
    return text 
   
```

```python
df['review']=df['review'].apply(rem_punc)
```

```python

from nltk.corpus import stopwords

# Download stopwords if not already downloaded
import nltk
nltk.download('stopwords')

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))  # Move the stopwords set creation inside the function
    
    new_text = []
    for word in text.split():
        if word not in stop_words:  # Check if the word is not in the stopwords set
            new_text.append(word)
    print(new_text)
    x=new_text[ : ]
    new_text.clear()
    return " ".join(x)  # Join the list of words into a single string
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

```python
df['review']
```

```python
#separating x and y

x=df.iloc[:,0:1]
y=df['sentiment']
```

```python
x
```

```python
y
```

```python
#converting y column to 0 and 1
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y=encoder.fit_transform(y)
```

```python
print(y)
```

```python
#split train test data

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
# applying BOW 
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()



X_train_bow=cv.fit_transform(X_train['review']).toarray()
X_test_bow= cv.transform(X_test['review']).toarray()
```

```python
X_train_bow.shape
```

```python
# applying naive bayes


from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()    # using guassian naive bayes

gnb.fit(X_train_bow,y_train)
```

```python
y_pred = gnb.predict(X_test_bow)

from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_pred)
```

Accuracy is poor as we are using limited data because of processor constrain

```python
confusion_matrix(y_test,y_pred)
```

```python
# applying random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

rf.fit(X_train_bow,y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

accuracy improved by using random forest from 0.54 to 0.8

```python
cv = CountVectorizer(max_features=3000)  # taking only frequently used 3000 words

X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

rf = RandomForestClassifier()

rf.fit(X_train_bow,y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```

```python
cv = CountVectorizer(ngram_range=(1,2),max_features=5000)    # using ngrams


X_train_bow = cv.fit_transform(X_train['review']).toarray()
X_test_bow = cv.transform(X_test['review']).toarray()

rf = RandomForestClassifier()

rf.fit(X_train_bow,y_train)
y_pred = rf.predict(X_test_bow)
accuracy_score(y_test,y_pred)
```
